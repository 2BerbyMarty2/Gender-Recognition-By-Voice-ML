# Naive Bayes Classifier

## Overview

Naive Bayes is a supervised machine learning algorithm used for classification tasks. It is based on Bayes' Theorem and predicts the probability that a data point belongs to a particular class.

The algorithm is called *naive* because it assumes that all features are independent of one another, even though this assumption is rarely true in real-world datasets.

## How It Works

Given a set of features:

```text
X = (x₁, x₂, x₃, ..., xₙ)
```

Naive Bayes calculates the probability that a sample belongs to each class and assigns the sample to the class with the highest probability.

For example, in a voice classification task:

```text
P(Male | Features)
P(Female | Features)
```

The class with the larger probability becomes the prediction.

## Bayes' Theorem

The algorithm is based on Bayes' Theorem:

```math
P(A|B) = \frac{P(B|A)P(A)}{P(B)}
```

Where:

* **P(A|B)** – Posterior Probability
* **P(B|A)** – Likelihood
* **P(A)** – Prior Probability
* **P(B)** – Evidence

## Why It Is Called "Naive"

Naive Bayes assumes that all features are independent of one another.

For a voice classification dataset, this means assuming that:

* RMS Energy
* Spectral Centroid
* Spectral Bandwidth

are independent features.

Although this assumption is often unrealistic, the model can still perform surprisingly well in practice.

## Training Process

1. Calculate the probability of each class.
2. Learn the feature distributions for each class.
3. Store the learned probabilities.
4. No iterative optimization is required.

## Prediction Process

For a new sample:

1. Calculate the probability for each class.
2. Compare the probabilities.
3. Select the class with the highest probability.

## Advantages

* Very fast training
* Very fast prediction
* Works well on small datasets
* Easy to understand and implement
* Requires little computational power
* Provides a strong baseline model

## Disadvantages

* Assumes feature independence
* May perform poorly when features are highly correlated
* Usually less accurate than advanced models such as SVM, Random Forest, or XGBoost

## Types of Naive Bayes

### Gaussian Naive Bayes

Used for continuous numerical features and assumes features follow a Gaussian (normal) distribution.

### Multinomial Naive Bayes

Commonly used for text classification and word-frequency data.

### Bernoulli Naive Bayes

Used when features are binary, such as true/false or yes/no values.

## Application to Voice Classification

Since the voice dataset contains continuous numerical features such as RMS, spectral centroid, and spectral bandwidth, **Gaussian Naive Bayes (GaussianNB)** is the most appropriate variant.

The model learns the distribution of each feature for male and female voices and uses these distributions to estimate the probability that a new voice sample belongs to either class.

## Comparison with SVM

| Naive Bayes                  | SVM                        |
| ---------------------------- | -------------------------- |
| Probabilistic model          | Margin-based model         |
| Very fast training           | Slower training            |
| Assumes feature independence | No independence assumption |
| Easy to interpret            | More complex               |
| Good baseline model          | Often higher accuracy      |

## Conclusion

Gaussian Naive Bayes is a simple, efficient, and effective classification algorithm that serves as an excellent baseline for voice classification tasks. While its independence assumption can limit performance, it provides fast training and prediction while offering a useful benchmark against more advanced models such as SVM, Random Forest, and XGBoost.
